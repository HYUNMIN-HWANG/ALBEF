coco2017 데이터로 json 파일 만들기

```python
import os
import json

with open("D:\\Data\\COCO\\annotations\\captions_val2017.json", 'r') as f :
    st_python = json.load(f)
st_python['images'][0]
```

```python
st_python['annotations'][0]
```

```python
from collections import OrderedDict
root = 'D:\\Data\\COCO\\val2017\\'
annot_list = []

for i, row in enumerate(st_python['images']) : 
    file_data = {}
    image = root + row['file_name']
    file_data['image_id'] = row['id']
    file_data['image'] = image
    annot_list.append(file_data)
```

```python
annot_list[0]
```

```python
for i, row in enumerate(st_python['annotations']) : 
    for j, row1 in enumerate(annot_list) : 
        if row['image_id'] == row1['image_id'] :
            annot_list[j]['caption'] = row['caption']
```

```python
annot_list[0]
```

```python
len(annot_list)
```

```python
with open("D:\ALBEF_dir\json_pretrain\captions_val2017_re.json", 'w', encoding='utf-8') as make_file:
    json.dump(annot_list, make_file, indent="\t")
```
